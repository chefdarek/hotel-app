```python
import pandas as pd
import datetime
import random
import decimal
import numpy as np
import matplotlib.pyplot as plt
```

```python
Fs = 365
f = 1
sample = 365
x = np.arange(sample)
sine = np.sin(2 * np.pi * f * x / Fs)
Fs = 365
fx = 3
fy = 10
fz = 52
fq = 24
sample = 365
x = np.arange(sample)
sine2 = np.sin(2 * np.pi * fx * x / Fs)
sine3 = np.sin(2 * np.pi * fy * x / Fs)
sine4 = np.sin(2 * np.pi * fz * x / Fs)
sine5 = np.sin(2 * np.pi * fq * x / Fs)
random_decimals = np.arange(3,12,.33)
start_date = datetime.date(2020, 1, 1)
days = pd.date_range(start_date, periods=365, freq='D')
totrooms = [81]*365
gop = np.arange(20,30,0.1)
gop_list = []
for x in range(365): gop_list.append(np.random.choice(gop))
market_adr = [.10,-.05, .11, -.11, .08, -.08, .06, -.06]
mark_adr_list = []
for x in range(365): mark_adr_list.append(np.random.choice(market_adr))
market_basket = [75,68,90,92,65,67,73,55,45,30,87,72]
mark_list = []
for x in range(365): mark_list.append(np.random.choice(market_basket)/100)



mpcb_choices = np.arange(1,50,1)
mpcb_list = []
for x in range(365): mpcb_list.append(np.random.choice(mpcb_choices)/100)
sent_choices = np.arange(75,100,.2)
sent_list = []
for x in range(365): sent_list.append(np.random.choice(sent_choices)/100)
drr_choices = np.arange(.05,.40,.01)
drr_list = []
for x in range(365): drr_list.append(np.random.choice(drr_choices))
web_choices = np.arange(.005,.05,.01)
web_list = []
for x in range(365): web_list.append(np.random.choice(web_choices))
seg_choices = np.arange(.005,.05,.01)
seg_list = []

tturn = np.arange(40,70,.22)
tturn_list = []
for x in range(365): tturn_list.append(np.random.choice(tturn))
    
speed_average = list(np.arange(7,25,.15))
speed_average = list([x+(x%3) for x in tturn_list])
speed_average_list = []
for x in range(365): speed_average_list.append(np.random.choice(speed_average))
speed_average_list.sort()
delivery = list([x+np.random.choice(random_decimals) for x in speed_average_list])
delivery.sort()
music_dec_list = []
reservation = np.arange(100,350,1)
reservation_list = []
for x in range(365): reservation_list.append(np.random.choice(reservation))
reservation_list.sort()
for x in range(365): music_dec_list.append(np.random.choice([55,75,35]))
music_dec_list.sort()

mac_dwell_s = list([x+np.random.choice(random_decimals) for x in tturn_list])
mac_dwell_s.sort()
lumen_list = []
for x in range(365): lumen_list.append(np.random.choice([150,200,250,300,350]))
lumen_list.sort(reverse=True)
menu_browse = list([x/np.random.choice(random_decimals) for x in speed_average_list])
menu_browse.sort(reverse=True)
#segments group, transient, corporate, packages, foriegn 
# for x in range(365): seg_list.append([random.choice(seg_choices),
#                                       random.choice(seg_choices),
#                                       random.choice(seg_choices),
#                                       random.choice(seg_choices),
#                                       random.choice(seg_choices)])
trev_choices = np.arange(.10,.30,.01)
trev_list = []
for x in range(365): trev_list.append(random.choice(trev_choices))

df = pd.DataFrame(columns=['timestamp','Total Rooms','Out of Service','Available Rooms','Occupied Rooms',
                           'Occupied Percent','Room Revenue','Average Daily Rate ADR','market_adr_raw','Market ADR', 'Revenue Per Available Room RevPAR',
                           'Market Rev Difference','Market RevPar','Average Occupancy Rate AOR','Reservations','Music Decibles','Light Lumens',
                           'Avg Table Turn Time Minutes', 'Plate Speed Average RFID','Beverage Speed Average RFID',
                           'Take Out Food Average Hold Time','Third Party Delivery Times','MAC Address Dwell', 'Average Guest Age',
                           'Facial Sentiment Score','Guest Recognition','Menu Browse Minutes','Food COGS %', 'Beverage COGS %',
                           'Gross Operating Profit per Available Room GOPPAR','market_perc','Market Penetration Index MPI','Average Rate Index','Revenue Generation Index RGI',
                           'Market Cost per Booking','Sentiment Analysis','Direct Revenue Ratio DRR','Web Conversion','Segmentation',
                           'outlet_net_perc','Outlet Net Revenue','Total Revenue','Total Revenue per Available Room TREVPAR']);
```

```python
#mock data
market_rev_b = [75,-68,90,-92,65,67,110,-55,100,125,109,
                99,122,121,-45,-30,87,72]
mark_rev = []
for x in range(365): mark_rev.append(np.random.choice(market_rev_b))
df['Market Rev Difference'] = list(mark_rev)
df['Market Rev Difference'].head()
```

```python
df['timestamp'] = days
df['Total Rooms'] = totrooms
df['Out of Service'] = (sine3 + 3) - (sine3 + 3)%3
df['Available Rooms'] = df['Total Rooms'] - df['Out of Service']
df['Occupied Rooms'] = df['Available Rooms'] + sine4
df['Occupied Percent'] = round(df['Occupied Rooms'] / df['Total Rooms'], 4)
df['Room Revenue'] = sine4 + 11786
df['Average Daily Rate ADR'] = sine+150
df['market_adr_raw'] = sine2 + 100
df['Market ADR'] = (df['Average Daily Rate ADR'] * df['market_adr_raw'])+ df['Average Daily Rate ADR']
df['Revenue Per Available Room RevPAR'] = round(df['Room Revenue']/df['Available Rooms'])

df['Market RevPar']= df['Revenue Per Available Room RevPAR'] + df['Market Rev Difference']
df['Food COGS %'] = round((df['Market ADR']/1000) + 10,2)
df['Beverage COGS %'] = round(df['Market ADR']/1000 +3.65,2)
df['Reservations'] = reservation_list
df['Music Decibles'] = music_dec_list
df['Light Lumens'] = lumen_list
tturn_list.sort(reverse=True)
df['Avg Table Turn Time Minutes'] = tturn_list
tturn_list.sort(reverse=True)
df['Plate Speed Average RFID'] = speed_average_list
df['Beverage Speed Average RFID'] =speed_average_list
speed_average_list.sort(reverse=True)
df['Take Out Food Average Hold Time'] = speed_average_list
df['Third Party Delivery Times'] = delivery
df['MAC Address Dwell'] = mac_dwell_s
df['Average Guest Age'] = sine2 + 30
sent_list.sort()
df['Facial Sentiment Score'] = sent_list
df['Guest Recognition'] = sent_list
df['Menu Browse Minutes'] = menu_browse

df['Average Occupancy Rate AOR'] = round(df['Occupied Rooms']/df['Available Rooms'],2)
df['Gross Operating Profit per Available Room GOPPAR'] = sine5 + 23.87
df['market_perc'] = mark_list
df['Market Penetration Index MPI'] = df['Occupied Percent'] / df['market_perc']
# adr_df = pd.DataFrame(columns=['timestamp','room_num','rate','rate_type'])
df['Average Rate Index'] = df['Average Daily Rate ADR']/df['Market ADR']
df['Revenue Generation Index RGI'] = df['Revenue Per Available Room RevPAR']/df['Market RevPar']
df['Market Cost per Booking'] = mpcb_list
df['Sentiment Analysis'] = sent_list
df['Direct Revenue Ratio DRR'] = drr_list
df['Web Conversion'] = web_list
#segments group, transient, corporate, packages, foriegn
#df['Segmentation'] = seg_list
df['outlet_net_perc'] = trev_list
df['Outlet Net Revenue'] = (df['Room Revenue'] * df['outlet_net_perc'])
df['Total Revenue'] = df['Room Revenue'] + df['Outlet Net Revenue']
df['Total Revenue per Available Room TREVPAR'] = df['Total Revenue']/ df['Available Rooms']

df.sample(20)
```

```python
these = df.columns.to_list()
these.sort()
print(these)
```

```python
df = df.drop(['market_adr_raw','market_perc','outlet_net_perc'], axis=1)
monthly = df[df['timestamp'].dt.day == 1]
```

```python
df.to_csv('hotel_metric.csv',index=False)
monthly.to_csv('monthly.csv', index=False)
```
